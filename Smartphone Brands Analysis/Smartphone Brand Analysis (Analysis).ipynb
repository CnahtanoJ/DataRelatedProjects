#### Memanfaatkan valuecount

```python
def valuecount(df,Nama_Kolom):
    list_a=[]
    lol=1
    for j in df[Nama_Kolom]:
        if pd.isna(j):
            lol=1
        else:
            list_a.extend(j.split(','))
    final = pd.DataFrame(list_a).value_counts(normalize=False)
    return final
```
